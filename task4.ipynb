```python
import tensorflow as tf
import keras
from keras.models import Sequential
import numpy as np
```

```python
model=Sequential()
model_part1=tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_tensor=None,
    input_shape=(32,32,3),
    pooling=None,
    classes=10,
)
```

```python
for layer in model_part1.layers:
    layer.trainable=False
```

```python
model.add(model_part1)
```

```python
from keras.layers import Flatten,Dense
```

```python

model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
fashion_mnist=keras.datasets.fashion_mnist
```

```python
(X_train,Y_train),(X_test,Y_test)=fashion_mnist.load_data()
```

```python
X_train.shape
```

```python
X_train = np.expand_dims(X_train, axis=-1)
X_train = tf.image.resize(X_train, [32,32]) 
```

```python
X_train.shape
```

```python
X_test = np.expand_dims(X_test, axis=-1)
X_test = tf.image.resize(X_test, [32,32])
```

```python
X_test.shape
```

```python
train_files=[]
```

```python
import cv2
```

```python
for i,fi in enumerate(X_train):
    
    im_array=np.array(fi)
    img= cv2.cvtColor(im_array,cv2.COLOR_GRAY2RGB)
    train_files.append(img)
```

```python
test_files=[]
```

```python
for i,fi in enumerate(X_test):
    im_array=np.array(fi)
    img= cv2.cvtColor(im_array,cv2.COLOR_GRAY2RGB)
    test_files.append(img)
```

```python
X1_train=np.zeros((len(train_files),32,32,3))
for i,fi in enumerate(train_files):
    fi_array=np.array(fi)
    if(fi_array.shape[0]==32 and fi_array.shape[1]==32):
        X1_train[i,:,:,:]=fi_array
```

```python
X1_test=np.zeros((len(test_files),32,32,3))
for i,fi in enumerate(test_files):
    fi_array=np.array(fi)
    if(fi_array.shape[0]==32 and fi_array.shape[1]==32):
        X1_test[i,:,:,:]=fi_array
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate= 1e-3),
loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
```

```python
model.fit(X1_train, Y_train, epochs=10, batch_size=15, validation_data=(X1_test, Y_test))
```
